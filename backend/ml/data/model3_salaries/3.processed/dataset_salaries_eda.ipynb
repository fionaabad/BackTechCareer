```python
import pandas as pd
```

```python
df = pd.read_csv("salaries_unified.csv")
```

```python
print("Shape:", df.shape)
print("\nTop 10 países por nº de filas:")
print(df["country"].value_counts().head(10))

print("\nSalario global (USD) - describe:")
print(df["salary_in_usd"].describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.95]))

print("\nSalario por seniority:")
print(
    df.groupby("seniority")["salary_in_usd"]
      .describe(percentiles=[0.25, 0.5, 0.75])
)

print("\nSalario por role_label (top 5 roles):")
top_roles = df["role_label"].value_counts().head(5).index.tolist()
print(
    df[df["role_label"].isin(top_roles)]
      .groupby("role_label")["salary_in_usd"]
      .describe(percentiles=[0.25, 0.5, 0.75])
)
```
